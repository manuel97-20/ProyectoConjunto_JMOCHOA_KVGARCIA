*El siguiente, es un algoritmo el cual muestra los numeros primos en el intervalo 2 - 100

```python

##Numero primos

for i in range(2,100):
    c = 2
    primo = True
    while primo and c<i:
        if i%c == 0:
            primo = False
        else:
            c+=1
    if primo:
        print(i)

```

```python
##fibonacci
n1 = 1
n2 = 1
print (n1)
print (n2)

count = 1
while count < 99:
    f = n1 + n2
    n1 = n2
    n2 = f
    count += 1
    print(f)
    
```
